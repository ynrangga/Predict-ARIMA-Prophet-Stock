```python
import yfinance as yf
import pandas as pd

# Tentukan tanggal mulai dan akhir
start_date = "2019-01-01"
end_date = "2025-04-29"

# Ambil data saham rentang waktu tertentu
hist = yf.Ticker('INDF.JK')
df = hist.history(start=start_date, end=end_date, auto_adjust=True)
df

# Menghapus zona waktu
df.index = df.index.tz_localize(None)

# Hapus kolom Dividends dan Stock Splits
df = df.drop(columns=['Dividends', 'Stock Splits'])

# Ubah index menjadi kolom Date dengan format tanggal saja
df = df.reset_index()
df['Date'] = df['Date'].dt.date

# Membulatkan angka desimal menjadi 2 angka di belakang koma
df[['Open', 'High', 'Low', 'Close']] = df[['Open', 'High', 'Low', 'Close']].round(2)

# Menampilkan data
print(df.head(4))
print('....')
print(df.tail(4))
```

```python
# Menyimpan data ke dalam file
excel_filename = 'data_saham_INDF.xlsx'
df.to_excel(excel_filename, index=False)
```
